# Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data into a pandas DataFrame
data = pd.read_csv('Tesla stock price prediction.csv')
```

```python
# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])
```

# Plotting the stock prices over time 

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Tesla Stock Prices Over Time')
plt.xticks(rotation=45)
plt.show()
```

# Plotting the distribution of the closing price 

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Close'], kde=True)
plt.xlabel('Closing Price')
plt.ylabel('Count')
plt.title('Distribution of Tesla Closing Prices')
plt.show()
```

# Plotting the moving average over time 

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Close'], label='Closing Price')
plt.plot(data['Date'], data['Close'].rolling(window=30).mean(), label='30-day Moving Average')
plt.plot(data['Date'], data['Close'].rolling(window=50).mean(), label='50-day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Tesla Stock Prices with Moving Averages')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

# Plotting the volume of trades over time 

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Volume'])
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Tesla Stock Volume over Time')
plt.xticks(rotation=45)
plt.show()
```

# Plotting candlestick chart for a specific time period 

```python
start_date = '2023-01-01'
end_date = '2023-06-30'
subset = data[(data['Date'] >= start_date) & (data['Date'] <= end_date)]
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(subset['Date'], subset['Close'], color='black')
ax.fill_between(subset['Date'], subset['Open'], subset['Close'], where=subset['Close'] >= subset['Open'], facecolor='green', alpha=0.5)
ax.fill_between(subset['Date'], subset['Open'], subset['Close'], where=subset['Close'] < subset['Open'], facecolor='red', alpha=0.5)
ax.xaxis.set_major_locator(plt.MaxNLocator(6))
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.set_title('Tesla Stock Price Candlestick Chart')
plt.xticks(rotation=45)
plt.show()
```
